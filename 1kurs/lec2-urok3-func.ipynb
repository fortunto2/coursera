```python
# coding: utf-8
import json
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
fx = np.arange(1, 16, 0.1)
# print x

fy = [f(i) for i in fx]
# print y
```

```python
# a*1 + b = f(1) 
# a*15 + b = f(15)

mnogo=np.array([[1,1],[15,1]])
print mnogo

vektor=np.array([[f(1)],[f(15)]])
print vektor

solve = np.linalg.solve(mnogo,vektor)

y = [i*solve[0][0]+solve[1][0] for i in x]

plt.plot(fx,fy,fx,y)
plt.show()
print solve
```

```python
# 1 x A + 1 x B + c = f(1)
# 8^2 x A + 8 x B + c = f(8)
# 15^2 x A + 15 x B + c = f(15)

mnogo=np.array([[1,1,1],[8**2,8,1],[15**2,15,1]])
print mnogo

vektor=np.array([[f(1)],[f(8)],[f(15)]])
print vektor

solve = np.linalg.solve(mnogo,vektor)

y = [(i**2)*solve[0][0]+i*solve[0][0]+solve[1][0] for i in x]

plt.plot(fx,fy,fx,y)
plt.show()
print solve
```
